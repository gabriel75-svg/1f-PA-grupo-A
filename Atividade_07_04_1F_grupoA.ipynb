```python
# Escrever um programa que recebe o nome e a idade de usuário e verifica se o usuário é menor de idade, adulto e idoso

# >= 17 é menor de idade
# =< 59 é adulto
# idoso

# Entrada

nome = str (input("Digite seu nome: "))
idade = int (input("Digite sua idade: "))

# Processamento/Saída

if idade <= 17:
    print("Você é um menor de idade")
elif idade <= 59:
    print("Você é um adulto")
else:
    print("Você é um idoso")

```

```python
# Escreva um programa que leia o valor de uma compra e aplique o desconto conforme as regras, Até R$100: sem desconto, de R$101 a R$500: 10% de desconto e acima de R$500: 20% de desconto

Valor = float(input("Digite o valor da compra: "))

# Processamento/Saída

if Valor <= 100:
    print("Você não tem desconto")
if Valor >= 101 and Valor <= 500:
    print("Você tem 10% de desconto")
    print(f"O valor com desconto é {Valor * 0.9}")
if Valor >= 501:
    print("Você tem 20% de desconto")
    print(f"O valor com desconto é {Valor * 0.8}")

```

```python
# Escrever um programa que classifica os tipos de triângulo e valida se é um triângulo.

# Validar se lados digitados pode formar um triângulo
# lado1 + lado2 > lado3 and lado1 + lado3 > lado2 and lado2 + lado3 > lado1

# Tipos de triângulo
# Equilatero - todos os lados iguais - ==
# Escaleno - todos os lados diferentes - !=
# Isóceles - dois lados iguais

# Entrada



lado1 = int(input("Digite o segundo lado do triângulo: "))
lado2 = int(input("Digite o segundo lado do triângulo: "))
lado3 = int(input("Digite o segundo lado do triângulo: "))

# Verifique se o triângulo é válido

# Processamento/Saída

if lado1 + lado2 > lado3 and lado1 + lado3 > lado2 and lado2 + lado3 > lado1:
    # Se for um triângulo válido, então classifique o tipo



    if lado1 == lado2 and lado2 == lado3 and lado3 == lado1:
        print(f"Os lados {lado1}, {lado2} e {lado3} formam um triângulo Equilatero")



    elif lado1 != lado2 and lado1 != lado3 and lado2 != lado3:
        print(f"Os lados {lado1}, {lado2} e {lado3} formam um triângulo Escaleno")


    else:
        print(f"Os lados {lado1}, {lado2} e {lado3} formam um triângulo Isóceles")


else:
    # Se não for um triângulo válido
    print("Não é um triângulo válido.")
```

```python
# Escrever um progama que classifica o sistema de energia

# Entrada

Energia = int(input("Digite a quantidade de energia:"))

# Processamento/Saída

if Energia <= 0:
  print(f"Seu personagem está {Energia} energia, ele está Morto")

if Energia >= 1 and Energia <= 30:
  print(f"Seu personagem está {Energia} energia, ele está Crítico")

if Energia >= 31 and Energia <= 70:
  print(f"Seu personagem está {Energia} energia, ele está Normal")

if Energia >= 71 and Energia <= 100:
  print(f"Seu personagem está {Energia} energia, ele está Saudável")

```

```python
# Escrever um programa que mostra a média escolar e classifica se foi reprovado, recuperação e aprovado

# Entrada

nome = str(input("Digite seu nome: "))
nota1 = float(input("Digite sua nota1: "))
nota2 = float(input("Digite sua nota2: "))
nota3 = float(input("Digite sua nota3: "))
nota4 = float(input("Digite sua nota4: "))

# Processamento

media = (nota1 + nota2 + nota3 + nota4) / 4

# Saída

print (f"Sua média é {media}")

if media < 5:
  print("Você foi reprovado")

if media >= 5 and media <= 6.9:
  print("Você está de recuperação")

if media >= 7:
  print("Você foi aprovado")
```

```python
# Escreva um programa que leia o salário de um funcionário e sua pontuação de desempenho (menor que 5 sem bônus, 5 a 7 5% de bônus, 8 a 9 10% de bõnus, 10 15% de bônus.) bônus baseado no salário

# Entrada

Salario = float(input("Digite seu salário: "))
Desempenho = int(input("Digite sua pontuação de desempenho: "))

# Processamento

percentual_bonus = 0

if Desempenho < 5:
    percentual_bonus = 0
elif Desempenho >= 5 and Desempenho <= 7:
    percentual_bonus = 0.05  # 5%
elif Desempenho >= 8 and Desempenho <= 9:
    percentual_bonus = 0.10  # 10%
elif Desempenho == 10:
    percentual_bonus = 0.15  # 15%
else:
    print("Pontuação de desempenho inválida. Nenhum bônus aplicado.")

novo_salario = Salario + (Salario * percentual_bonus)

# Saída

print(f"Seu novo salário é {novo_salario}")
```
