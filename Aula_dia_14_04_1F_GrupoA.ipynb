```python
# Escreva um número ao usuário e informe se ele é par ou ímpar.

# Entrada

numero = int(input("Digite um número: "))

# Processamento/Saída

if numero % 2 == 0:
  print("O número é par")
else:
  print("O número é ímpar")
```

```python
# Escreva um programa onde coloca dois números e mostre qual é o maior (ou se são iguais).

# Entrada

numero1 = int(input("Digite o primeiro número: "))
numero2 = int(input("Digite o segundo número: "))

# Processamento/Saída

if numero1 > numero2:
  print("O primeiro número é maior")
else:
  if numero1 < numero2:
    print("O segundo número é maior")
```

```python
# Escreva um programa onde aparece se o aluno foi aprovado ou não

# Entrada

nota = float(input("Digite a nota do aluno: "))

# Processamento/Saída

if nota >= 7:
  print("O aluno foi aprovado")

if nota >= 5 and nota < 7:
  print("O aluno está em recuperação")

if nota < 5:
  print("O aluno foi reprovado")
```

```python
# Escreva um programa que classifica a idade da pessoa.

# Entrada

idade = int(input("Digite a idade da pessoa: "))

# Processamento/Saída

if idade <= 12:
  print("Criança")
elif idade >= 13 and idade <= 17:
  print("adolescente")
elif idade >= 18 and idade <= 59:
  print("Adulto")
else:
  print("Idoso")
```

```python
# Escreva um programa e peça um número de 1 a 7 e mostre o dia que representa na semana (ex: 1 = segunda).

# Entrada

numero = int(input("Digite um número de 1 a 7: "))

# Processamento/Saída

if numero == 1:
  print("Segunda")
elif numero == 2:
  print("Terça")
elif numero == 3:
  print("Quarta")
elif numero == 4:
  print("Quinta")
elif numero == 5:
  print("Sexta")
elif numero == 6:
  print("Sábado")
elif numero == 7:
  print("Domingo")
```

```python
# Escreava um programa e mostre um ano e informe se ele é bissexto ou não.

# Entrada

ano = int(input("Digite um ano: "))

# Processamento/Saída

if ano % 4 == 0:
  print("O ano é bissexto")
else:
  print("O ano não é bissexto")
```

```python
# Escreva um programa e peça o valor da compra, conforme o valor tem descontos (≥ 100: 10% de desconto, ≥ 50 e < 100: 5% de desconto e < 50: sem desconto)

# Entrada

valor = float(input("Digite o valor da compra: "))

# Processamento/Saída

if valor >= 100:
  print("O valor com desconto é: ", valor * 0.9)
if valor >= 50 and valor < 100:
  print("O valor com desconto é: ", valor * 0.95)
if valor <= 50:
  print("Sem desconto")
```
