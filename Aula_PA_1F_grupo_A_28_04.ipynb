```python
# Escrever um programa que o usuário digita uma palavra e o programa verifica se a palavra é palindroma e imprime a palavra

# entrada
palavra = str (input("Digite uma palavra: "))

# processamento

invertida = ""
for i in (palavra):
  invertida = i + invertida

#saída

if invertida == palavra:
  print ("A palavra é palindroma")
else:
  print ("A palavra não é palindroma")
```
