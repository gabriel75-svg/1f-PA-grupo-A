```python
# Estrutura de decisões

# Escrever um programa que recebe o nome e a idade de usuário e verifica se o usuário é criança, adolescente e adulto

# >= 12 é criança
# =< 17 é adolescente
# Adulto

# Entrada

nome = str (input("Digite seu nome: "))
idade = int (input("Digite sua idade: "))


# Processamento/Saída

if idade <= 12:
    print(f"{nome}, Você tem {idade} anos de idade, você é uma criança")
elif idade <= 17:
    print(f"{nome}, Você tem {idade} anos de idade, você é um adolescente")
else:
    print(f"{nome}, Você tem {idade} anos de idade, você é um adulto")
```

```python
# Em um jogo, um personagem possui um variavel chamada energia. O programa  deve: Se a energia for maior que 70 - Forte, se a energia estiver entre 30 e 70 - Cansado, se a energia estiver menor que 30 - Fraco.

# Entrada

Energia = int(input("Digite a energia do seu personagem: "))

# Processamento/Saída

if Energia >=70:
  print(f"Seu personagem está com {Energia}. Ele está com a energia forte")
elif Energia >= 30:
  print(f"Seu personagem está com {Energia}. Ele está com a energia média")
else:
  print(f"Seu personagem está com {Energia}. Ele está com a energia fraca")
```

```python
# Escrever um programa que classifica os tipos de triângulo e valida se é um triângulo.

# Validar se lados digitados pode formar um triângulo
# lado1 + lado2 > lado3 and lado1 + lado3 > lado2 and lado2 + lado3 > lado1

# Tipos de triângulo
# Equilatero - todos os lados iguais - ==
# Escaleno - todos os lados diferentes - !=
# Isóceles - dois lados iguais

# Entrada



lado1 = int(input("Digite o segundo lado do triângulo: "))
lado2 = int(input("Digite o segundo lado do triângulo: "))
lado3 = int(input("Digite o segundo lado do triângulo: "))

# Verifique se o triângulo é válido

# Processamento/Saída

if lado1 + lado2 > lado3 and lado1 + lado3 > lado2 and lado2 + lado3 > lado1:
    # Se for um triângulo válido, então classifique o tipo



    if lado1 == lado2 and lado2 == lado3 and lado3 == lado1:
        print(f"Os lados {lado1}, {lado2} e {lado3} formam um triângulo Equilatero")



    elif lado1 != lado2 and lado1 != lado3 and lado2 != lado3:
        print(f"Os lados {lado1}, {lado2} e {lado3} formam um triângulo Escaleno")


    else:
        print(f"Os lados {lado1}, {lado2} e {lado3} formam um triângulo Isóceles")


else:
    # Se não for um triângulo válido
    print("Não é um triângulo válido.")
```
