```python
# Operadores aritiméticos
# Fazer um programa que calcula operações com 02 números

# Entrada

a = int (input("digite um número:"))
b = int (input("digite um número:"))


# Processamento

soma = a + b
subtracao = a - b
multiplicacao = a * b
divisao = a / b
div_int = a // b
resto = a % b
potencia = a ** b

# Saída

print ("")
print ("-------------- OPERADORES ARITIMÉTICOS-------------")
print ("")
print ("---------------SOMA-----------------------")
print ("")
print ("----------------SUBTRAÇÃO-----------------------")
print ("")
print ("---------------MULTIPLICAÇÃO-----------------------")
print ("")
print ("----------------DIVISÃO-----------------------")
print ("")
print ("------------------DIVISÃO INTEIRA---------------------")
print ("")
print ("-------------------RESTO DA DIVISÃO-------------------")
print ("")
print ("--------------------Resto da Divisão-------------------")
print ("")
print ("---------------------Potenciação------------------------")
print (f"O resultado da SOMA de {a} MAIS {b} é igual : {soma}")
print ("")
print (f"O resultado da SUBTRAÇÃO de {a} MENOS {b} é igual : {subtracao}")
print ("")
print (f"O resultado da MULTIPLICAÇÃO de {a} POR {b} é igual : {multiplicacao}")
print ("")
print (f"O resultado da DIVISÃO de {a} POR {b} é igual : {divisao}")
print ("")
print (f"o resultado da DIVISÃO INTEIRA de {a} POR {b} é igual : {div_int}")
print ("")
print (f"O resto da DIVISÃO de {a} POR {b} é igual : {resto}")
print ("")
print (f"O resultado da POTENCIAÇÃO de {a} POR {b} é igual : {potencia}")
print ("---------------------------------------------------")
```
